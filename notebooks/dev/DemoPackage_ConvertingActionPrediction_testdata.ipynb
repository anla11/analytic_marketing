## 3.1. Preprocessing for Customer Purchase Journey Netherlands Data

Binning the touchpoints into: 
   - Accomodations
   - TouroperatorCompetitor
   - Touroperatorbrand
   - Information
   - FlightTickets

```python
import pandas as pd
df = pd.read_csv('./TravelData.csv')
master = df.copy()
master['touch']=pd.cut(master['type_touch'],[0,3,6,9,12,15,22],right=False,labels = ['Accomodations','TouroperatorCompetitor','Touroperatorbrand','Information','FlightTickets','0'])
master = master[master.touch !='0']
master.to_csv (r'dataframe.csv', index = False, header=True)
master
```
